```python
import csv

def return_contents(file_name):
    with open(file_name) as infile:
        reader = csv.reader(infile)
        return list(reader)

data1 = return_contents('C:\\Users\\Azmery\\Documents\\CSUF\\ISDS 577\\Group Project\\Updated_OC2.csv')

print(data1)

combined = []
for row in data1:
    combined.extend(row)

with open('C:\\Users\\Azmery\\Documents\\CSUF\\ISDS 577\\Group Project\\Updated_OC2_out.csv', 'w', newline='') as outfile:
    writer = csv.writer(outfile)
    writer.writerow(combined)
```
